# 2. Matplotlib basics

## Make a simple plot

Mari buat plot dasar untuk mulai!

````python
height = [150,160,165,185]
weight = [70, 80, 90, 100]

# draw the plot
# plt.plot(x, y)
plt.plot(height, weight)
````

Kita meneruskan dua array sebagai argumen input ke metode **plot()** untuk menampilkan plot yang diperlukan. Di sini perhatikan bahwa larik pertama muncul pada sumbu x dan larik kedua muncul pada sumbu y plot.

#### Title, Labels, and Legends
- Sekarang plot pertama kita sudah siap, mari kita tambahkan judul, dan beri nama sumbu x dan sumbu y masing-masing menggunakan metode title(), xlabel() dan ylabel().


````python
# draw the plot
plt.plot(height,weight)

# add title
plt.title("Relationship between height and weight", pad=10)

# label x axis
plt.xlabel("Height", labelpad=20)

# label y axis
plt.ylabel("Weight")
````

````python
calories_burnt = [65, 75, 95, 99]

# draw the plot for calories burnt
plt.plot(calories_burnt)

# draw the plot for weight
plt.plot(weight)
````

- Menambahkan **legends** juga sederhana di matplotlib, Anda dapat menggunakan `legend()` yang menggunakan **labels** dan **loc** sebagai nama label dan lokasi legenda pada gambar sebagai paremeter.

https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html

````python
# draw the plot for calories burnt
plt.plot(calories_burnt)

# draw the plot for weight
plt.plot(weight)

# add legend in the lower right part of the figure
plt.legend(labels=['Calories Burnt', 'Weight'], loc='best')

# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html
````

- Perhatikan bahwa pada plot sebelumnya, kita tidak dapat memahami bahwa masing-masing nilai ini milik orang yang berbeda.
- Lihat sumbu X, dapatkah kita menambahkan label untuk menunjukkan bahwa masing-masing milik orang yang berbeda?
- Nilai berlabel pada sumbu apa pun dikenal sebagai **tick**.
- Anda dapat menggunakan `xticks` untuk mengubah lokasi setiap centang dan labelnya. Mari kita lihat ini dalam contoh

````python
# draw the plot
plt.plot(calories_burnt)
plt.plot(weight)

# add legend in the lower right part of the figure
plt.legend(labels=['Calories Burnt', 'Weight'], loc='best')

# set labels for each of these persons
plt.xticks(ticks=[0,1,2,3], labels=['p1', 'p2', 'p3', 'p4']);
````

#### Size, Colors, Markers and Line styles

- Anda juga dapat menentukan ukuran gambar menggunakan metode `figure()` dan meneruskan nilai sebagai tupel dengan panjang baris dan kolom ke argumen figsize.
- Nilai panjang dianggap dalam **inches**.

````python
# figure size in inches
plt.figure(figsize=(10,5))

# draw the plot
plt.plot(calories_burnt)
plt.plot(weight)

# add legend in the lower right part of the figure
plt.legend(labels=['Calories Burnt', 'Weight'], loc='lower right')

# set labels for each of these persons
plt.xticks(ticks=[0,1,2,3], labels=['p1', 'p2', 'p3', 'p4']);
````

- Dengan setiap argumen X dan Y, Anda juga dapat meneruskan argumen ketiga opsional dalam bentuk string yang menunjukkan warna dan jenis garis plot.
- Format default adalah `b-` yang berarti **solid blue line**. Pada gambar di bawah ini kita menggunakan `go` yang artinya **green circles**. Demikian juga, kita dapat membuat banyak kombinasi seperti itu untuk memformat plot kita.

https://matplotlib.org/stable/gallery/color/named_colors.html

````python
# draw the plot
plt.plot(calories_burnt)
plt.plot(weight, 'ro')

# add legend in the lower right part of the figure
plt.legend(labels=['Calories Burnt', 'Weight'], loc='best')

# set labels for each of these persons
plt.xticks(ticks=[0,1,2,3], labels=['p1', 'p2', 'p3', 'p4']);
````

- Kita juga dapat memplot beberapa set data dengan meneruskan beberapa set argumen sumbu X dan Y dalam metode `plot()` seperti yang ditunjukkan.

#### Figure and subplots

- Kita dapat menggunakan metode `subplots()` untuk menambahkan lebih dari satu plot dalam satu gambar.
- Metode `subplots()` membutuhkan dua argumen: mereka adalah **nrows, ncols**. Mereka menunjukkan jumlah baris, jumlah kolom masing-masing.
- Metode ini membuat dua objek: **figure** dan **axes** yang kita simpan dalam variabel `fig` dan `ax`.
- Anda memplot setiap gambar dengan menentukan posisinya menggunakan indeks baris dan indeks kolom. Mari kita lihat contoh di bawah ini:

https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html

````python
# create 2 plots
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

# plot on 0 row and 0 column
ax[0,0].plot(calories_burnt,'go')

# plot on 0 row and 1 column
ax[0,1].plot(weight)

# set titles for subplots
ax[0,0].set_title("Calories Burnt")
ax[0,1].set_title("Weight")

# set ticks for each of these persons
ax[0,0].set_xticks(ticks=[0,1,2,3]);
ax[0,1].set_xticks(ticks=[0,1,2,3]);

# set labels for each of these persons
ax[0,0].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4']);
ax[0,1].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4']);
````

- Perhatikan bahwa pada gambar di atas kita memiliki dua plot kosong, itu karena kita membuat 4 subplot (2 baris dan 2 kolom).
- Sebagai data scientist, akan ada saatnya Anda harus memiliki sumbu yang sama untuk semua subplot Anda. Anda dapat melakukannya dengan menggunakan parameter **sharex** dan **sharey** dari `subplot()`.

````python
# create 2 plots
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6,6), sharex=True, sharey=True)

# plot on 0 row and 0 column
ax[0].plot(calories_burnt,'go')

# plot on 0 row and 1 column
ax[1].plot(weight)

# set titles for subplots
ax[0].set_title("Calories Burnt")
ax[1].set_title("Weight")

# set ticks for each of these persons
ax[0].set_xticks(ticks=[0,1,2,3]);
ax[1].set_xticks(ticks=[0,1,2,3]);

# set labels for each of these persons
ax[0].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4']);
ax[1].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4']);
````

- Perhatikan pada plot di atas, sekarang sumbu x dan y hanya diberi label satu kali untuk masing-masing plot luar. Ini karena plot bagian dalam "berbagi" kedua sumbu.
- Selain itu, hanya ada **dua plot** karena kita mengurangi jumlah baris menjadi 1 dan kolom menjadi 2 di `subplot()`.

# 3. Line Chart

- Kita akan membuat diagram garis untuk menunjukkan **harga rata-rata per item**.
- Dengan beberapa kumpulan data, Anda mungkin ingin memahami perubahan dalam satu variabel sebagai fungsi waktu, atau variabel kontinu yang serupa.
- Di matplotlib, **line chart** adalah plot default saat menggunakan `plot()`.

````python
price_by_item = data_BM.groupby('Item_Type').Item_MRP.mean()

price_by_item
````

````python
# mean price based on item type
price_by_item = data_BM.groupby('Item_Type').Item_MRP.mean()[:10]

x = price_by_item.index.tolist()
y = price_by_item.values.tolist()

# set figure size
plt.figure(figsize=(14, 8))

# set title
plt.title('Mean price for each item type')

# set axis labels
plt.xlabel('Item Type')
plt.ylabel('Mean Price')

# set xticks
plt.xticks(labels=x, ticks=np.arange(len(x)), rotation=45)

plt.plot(x, y)
````

#  4. Bar Chart

- Misalkan kita ingin melihat **berapa rata-rata penjualan untuk setiap jenis outlet?**
- Bagan batang adalah jenis visualisasi sederhana lainnya yang digunakan untuk variabel kategoris.
- Anda dapat menggunakan `plt.bar()` sebagai ganti `plt.plot()` untuk membuat diagram batang.

````python
# sales by outlet size
sales_by_outlet_size = data_BM.groupby('Outlet_Size').Item_Outlet_Sales.mean()

# sort by sales
sales_by_outlet_size.sort_values(inplace=True)

x = sales_by_outlet_size.index.tolist()
y = sales_by_outlet_size.values.tolist()

# set figure size
plt.figure(figsize=(14, 8))

# set axis labels
plt.xlabel('Outlet Size')
plt.ylabel('Sales')

# set title
plt.title('Mean sales for each outlet type')

# set xticks
plt.xticks(labels=x, ticks=np.arange(len(x)))

plt.bar(x, y, color=['red', 'orange', 'magenta'])
````

# 5. Histogram

- **Distribusi Harga Barang**
- Histogram adalah jenis plot yang sangat umum ketika kita melihat data seperti tinggi dan berat badan, harga saham, waktu tunggu untuk pelanggan, dll yang sifatnya kontinu.
- Data histogram diplot dalam rentang terhadap frekuensinya.
- Histogram adalah grafik yang sangat umum muncul dalam probabilitas dan statistik dan membentuk dasar untuk berbagai distribusi seperti distribusi -normal, distribusi t, dll.

````python
# set figure size
plt.figure(figsize=(14, 8))

# title
plt.title('Item MRP (price) distribution')

# xlabel
plt.xlabel('Item_MRP')

# ylabel
plt.ylabel('Frequency')

# plot histogram
plt.hist(data_BM['Item_MRP'], bins=20, color='lightblue');
````

# 6. Box Plots

- **Distribusi penjualan**
- Box plots menunjukkan tiga nilai kuartil dari distribusi bersama dengan nilai ekstrim (outlier).
- 'Whisker' meluas ke titik-titik yang terletak dalam 1,5 IQR dari kuartil bawah dan atas, dan kemudian pengamatan yang berada di luar kisaran ini ditampilkan secara independen.
- Ini berarti bahwa setiap nilai dalam boxplot sesuai dengan pengamatan sebenarnya dalam data.
- Mari kita coba untuk memvisualisasikan distribusi Item_Outlet_Sales item.

````python
data = data_BM[['Item_Outlet_Sales']]

# set figure size
plt.figure(figsize=(14, 8))

# create outlier point shape
red_diamond = dict(markerfacecolor='r', marker='D')

# set title
plt.title('Item Sales distribution')

# make the boxplot
plt.boxplot(data.values, labels=['Item Sales'], flierprops=red_diamond);
````

- Kita juga dapat membuat  multiple boxplots untuk kolom yang berbeda dari kumpulan data Anda.
- Untuk memplot beberapa boxplot, Anda dapat menggunakan `subplots()` yang sama seperti yang kita lihat sebelumnya.
- Mari kita lihat distribusi Item_Weight, Item_MRP bersama

````python
data = data_BM[['Item_Weight', 'Item_MRP']]

# create outlier point shape
red_diamond = dict(markerfacecolor='r', marker='D')

# generate subplots
fig, ax = plt.subplots(figsize=(14, 8))

# make the boxplot
plt.boxplot(data.values, labels=['Item Weight', 'Item MRP (price)'], flierprops=red_diamond);
````

# 7. Violin Plots

- Distribusi kepadatan berat Item dan harga Item

````python
data = data_BM[['Item_Weight', 'Item_MRP']]

# generate subplots
fig, ax = plt.subplots(figsize=(14, 8))

# add labels to x axis
plt.xticks(ticks=[1,2], labels=['Item Weight', 'Item MRP'])

# make the violinplot
plt.violinplot(data.values);
````

# 8. Scatter Plots

- **Distribusi relatif berat item dan visibilitasnya**
- Ini menggambarkan distribusi dua variabel menggunakan awan titik, di mana setiap titik mewakili pengamatan dalam kumpulan data.
- Penggambaran ini memungkinkan mata untuk menyimpulkan sejumlah besar informasi tentang apakah ada hubungan yang berarti di antara mereka.

**CATATAN : Di sini, kita hanya akan menggunakan sebagian data untuk plot.**

````python
# set figure size
plt.figure(figsize=(14, 8))

# set label of axes
plt.xlabel('Item_Weight')
plt.ylabel('Item_Visibility')

# plot
plt.scatter(data_BM["Item_Weight"][:200], data_BM["Item_Visibility"][:200])
````

# 9. Bubble Plots

- **Distribusi penjualan relatif, harga barang, dan visibilitas barang**
- Mari kita buat scatter plot Item_Outlet_Sales dan Item_MRP dan buat **size** gelembung dengan kolom Item_Visibility.
- Plot ini memungkinkan kita memahami hubungan saling ketergantungan antara 3 variabel.

**CATATAN: Perhatikan bahwa kita hanya menggunakan subset data untuk plot.**

````python
# set figure size
plt.figure(figsize=(14, 8))

# set label of axes
plt.xlabel('Item_MRP')
plt.ylabel('Item_Outlet_Sales')

# set title
plt.title('Item Outlet Sales vs Item MRP (price)')

# plot
plt.scatter(data_BM["Item_MRP"][:100], data_BM["Item_Outlet_Sales"][:100], s=data_BM["Item_Visibility"][:100]*1000, c='red')
````
